Дальше логика такая: если, например для кластера 1 значение в sent == received в кластере 2 => это перевод битков из 1 в 2. Ну и далее по аналогии для каждого пункта

```python
print('1 -> 2: ',sum_12['sent'].sum())
```

Чтобы посчитать fee складываем сколько кластер принял от двух других и вычитаем сколько отправил. В конце складываем fee для всех 3 кластеров и получается 0, что в общем то логично

```python
fee_1 = sum_21['sent'].sum()+sum_01['sent'].sum()-sum_12['sent'].sum()-sum_10['sent'].sum()
fee_1
```

```python
fee_2 = sum_12['sent'].sum()+sum_02['sent'].sum()-sum_21['sent'].sum()-sum_20['sent'].sum()
fee_2
```

```python
fee_0 = sum_10['sent'].sum()+sum_20['sent'].sum()-sum_01['sent'].sum()-sum_02['sent'].sum()
fee_0
```

```python
fee_1+fee_2+fee_0
```

Получилось что fee для кластеров 0,2 < 0. значит они отправляли больше чем получали. В кластере 1 fee единственное положительное и большое, походу там майнеры сидят :)
